```python
##^& deleting the column
del HrtFailure_Prd['record']
##^& check null values
print("Null or Missing Values in the data:", HrtFailure_Prd.isnull().values.sum())
##^& check duplicates values
print("Repeated or Duplicate Rows in the data:", HrtFailure_Prd.duplicated().sum())
```

EDA (Exploratory Data Analysis)

```python
##%@ CountPlot of 'Heart Disease' (type)
sns.countplot(data=HrtFailure_Prd, palette='coolwarm', x='type')
plt.title('Class Distribution in Heart Failure Data')
plt.xlabel('Cardiac condition')
plt.ylabel('Quantity')
plt.show()
```

```python
#$#^ distribution in a histogram
sns.histplot(HrtFailure_Prd['0_pre-RR'], kde=True, color='purple')
plt.title('RR Intervals Distribution')
plt.xlabel('RR Intervals Features')
plt.ylabel('Frequency')
plt.show()
```

```python
##%$ scatter plot
sns.scatterplot(data=HrtFailure_Prd, x='0_qrs_interval', y='0_st_interval', hue='type', palette='inferno')
plt.show()
```

```python
#Convert object into numbering using label encoding

from sklearn.preprocessing import LabelEncoder as LabelEncoder

encoder = LabelEncoder()
HrtFailure_Prd['type'] = encoder.fit_transform(HrtFailure_Prd['type'])
HrtFailure_Prd
```
